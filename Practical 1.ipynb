```python
from pandas import read_csv
from matplotlib import pyplot
from pandas import DataFrame
from pandas import Grouper
series = read_csv('C:/16TSV/daily-min-temperatures.csv', header=0, index_col=0,parse_dates=True)
print(series.head())
series=series.squeeze()
type(series)
print(series.describe())
```

```python
series.plot()
pyplot.show()
```

```python
series.plot(style = 'b.')
pyplot.show()
```

```python
series.plot(style = 'g--')
pyplot.show()
```

```python

groups = series.groupby(Grouper(freq = 'A'))
print(groups)
years = DataFrame()
for name, group in groups:
  years[name.year] = group.values
  print(years)
years.plot(subplots=True, legend=False)
pyplot.show()
```

```python
series.hist()
pyplot.show()
```

```python
series.plot(kind='kde')
pyplot.show()
```

```python
years.boxplot()
pyplot.show()
```

```python
years.boxplot()
pyplot.show()
```

```python
years = years.T
print(years)
pyplot.matshow(years, interpolation=None, aspect='auto')
pyplot.show()
```

```python
from pandas.plotting import lag_plot
lag_plot(series)
pyplot.show()
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(series)
pyplot.show()
```
